```python
y=data["Performance_Score"]
X=data[['Years_At_Company','Monthly_Salary','Overtime_Hours','Promotions','Employee_Satisfaction_Score']]
```

```python
X
```

```python
X.columns
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
```

```python
from sklearn.model_selection import train_test_split
# X = features, y = target
# X, y should already be prepared

X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.2,)
```

```python


X_train = scaler.fit_transform(X_train)
```

```python
import joblib
joblib.dump(scaler,"scaler.pkl")
```

```python
X_train
```

```python
X_test=scaler.fit_transform(X_test)
```

```python
X_test
```

```python
import warnings 
warnings.filterwarnings("ignore")
```

```python
from sklearn.metrics import accuracy_score
```

```python
def modelperformance(predictions):
    print("Accuracy score on model is {}".format(accuracy_score(y_test,predictions)))

```

```python
from sklearn.model_selection import GridSearchCV
```

```python
from sklearn.linear_model import LogisticRegression
log_model=LogisticRegression()
```

```python
log_model.fit(X_train,y_train)
```

```python
log_model.predict([[0,0,0,0,0]])
```

```python
log_model.predict(X_test)
```

```python
log_prediction=log_model.predict(X_test)
```

```python
modelperformance(log_prediction)
```

```python
from sklearn.neighbors import KNeighborsClassifier
param_grid={"n_neighbors":[3,5,7,9,11],
            "weights":["uniform","distance"]}
            
```

```python
gridkn=GridSearchCV(KNeighborsClassifier(),param_grid,cv=5)
```

```python
gridkn.fit(X_train,y_train)
```

```python
gridkn.best_params_
```

```python
predicitons=gridkn.predict(X_test)
```

```python
modelperformance(predicitons)
```

```python
from sklearn.svm import SVC 
svm=SVC()
```

```python
param_grid={
    "C":[0.01,0.1,0.5,1],
    "kernel":["linear","rbf","poly"]
}
```

```python
gridsvc=GridSearchCV(svm,param_grid,cv=5)
```

```python
gridsvc.fit(X_train,y_train)
```

```python
gridsvc.predict(X_test)
```

```python
y_pred=gridsvc.predict(X_test)
```

```python
modelperformance(y_pred)
```

```python
from sklearn.tree import DecisionTreeClassifier

dt = DecisionTreeClassifier(random_state=42)
dt.fit(X_train, y_train)

y_pred= dt.predict(X_test)

modelperformance(y_pred)

```

```python
joblib.dump(dt, "decision_tree_model.pkl")
```
